```python
import pandas as pd
import numpy as np
df = pd.read_csv('framingham.csv')
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sns.countplot(x='TenYearCHD',hue='male',data=df)
```

```python
df['education'] = df['education'].fillna(value = df['education'].mean())
df['BPMeds'] = df['BPMeds'].fillna(value = df['BPMeds'].mean())
df['totChol'] = df['totChol'].fillna(value = df['totChol'].mean())
df['cigsPerDay'] = df['cigsPerDay'].fillna(value = df['cigsPerDay'].mean())
df['BMI'] = df['BMI'].fillna(value = df['BMI'].mean())
df['heartRate'] = df['heartRate'].fillna(value = df['heartRate'].mean())
df['glucose'] = df['glucose'].fillna(value = df['glucose'].mean())
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
x = df.iloc[:,0:15]
x
```

```python
y = df.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
from sklearn.linear_model import LogisticRegression 
import matplotlib.pyplot as plt
lr = LogisticRegression(solver='lbfgs', max_iter=10000)
lr = lr.fit(x_train,y_train.values.ravel())
perd = lr.predict(x_test)
perd.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test,perd))
print("\n")
print(classification_report(y_test,perd))
print("\n")
print(accuracy_score(y_test,perd)*100)
```

```python
Comparison = pd.DataFrame({'Actual':y_test["TenYearCHD"],'Predicted':perd})
Comparison.head(15)
```
